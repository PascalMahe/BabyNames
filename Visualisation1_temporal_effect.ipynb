```python
import altair as alt
import pandas as pd
import numpy as np
alt.data_transformers.enable("vegafusion")
names = pd.read_csv("dpt2020.csv", sep=";")
names.sample(5)
names.drop(names[names.preusuel == '_PRENOMS_RARES'].index, inplace=True)
names.drop(names[names.dpt == 'XX'].index, inplace=True)
```

```python
names.sample(5)
```

```python

#sum all the occurence independantly of the departement
#names = names.groupby(['preusuel', 'annais','sexe'])['nombre'].sum().reset_index()
names = names.groupby(['preusuel', 'annais'])['nombre'].sum().reset_index()
#gender renaming
#names['sexe'] = names['sexe'].replace(1, 'male')
#names['sexe'] = names['sexe'].replace(2, 'female')

#lowercase
names['preusuel'] = names['preusuel'].str.lower()
```

```python
names.sample(5)
```

```python
# Filter for demonstration
names_copy = names
```

```python
from IPython.display import display,clear_output
import ipywidgets as widgets

names = names_copy.loc[names_copy['preusuel'].isin(["jean"])]
searchbar = widgets.Text(
    value='jean',
    placeholder='Enter names (use ";" as a separator):',
    description='Names',
    disabled=False   
)

display(searchbar)

# Create a selection that chooses the nearest point & selects based on x-value
nearest = alt.selection_point(nearest=True, on="pointerover",
                                fields=["annais"], empty=False)



line = alt.Chart(names).mark_line().encode(
    x=alt.X('annais:T', axis=alt.Axis(
        format="%Y", tickCount=20, title='Year', labelAngle=45)),
    y="nombre:Q",
    color="preusuel:N"
)
# Transparent selectors across the chart. This is what tells us
# the x-value of the cursor
selectors = alt.Chart(names).mark_point().encode(
    x='annais:T',
    opacity=alt.value(0),
).add_params(
    nearest
)
# Draw points on the line, and highlight based on selection
points = line.mark_point().encode(
    opacity=alt.condition(nearest, alt.value(1), alt.value(0))
)
# Draw text labels near the points, and highlight based on selection
text = line.mark_text(align="left", dx=5, dy=-5).encode(
    text=alt.condition(nearest, "nombre:Q", alt.value(" "))
)
# Draw a rule at the location of the selection
rules = alt.Chart(names).mark_rule(color="gray").encode(
    x='annais:T'
).transform_filter(
    nearest
)
# Put the five layers into a chart and bind the data
layers = alt.layer(
    line, selectors, points, rules, text
).properties(
    width=600, height=300
)
display(layers)







def searchbar_callback(change):
    #names = names.loc[names['preusuel'].isin(names_filter)]
    filer_list = searchbar.value.split(";")
    names = names_copy
    names = names.loc[names['preusuel'].isin(filer_list)]
    print(filer_list)
    clear_output(wait=True)
    
    display(searchbar)
    
    # Create a selection that chooses the nearest point & selects based on x-value
    nearest = alt.selection_point(nearest=True, on="pointerover",
                                fields=["annais"], empty=False)



    line = alt.Chart(names).mark_line().encode(
        x=alt.X('annais:T', axis=alt.Axis(
            format="%Y", tickCount=120, title='Year', labelAngle=45)),
        y="nombre:Q",
        color="preusuel:N"
    )

    # Transparent selectors across the chart. This is what tells us
    # the x-value of the cursor
    selectors = alt.Chart(names).mark_point().encode(
        x='annais:T',
        opacity=alt.value(0),
    ).add_params(
        nearest
    )

    # Draw points on the line, and highlight based on selection
    points = line.mark_point().encode(
        opacity=alt.condition(nearest, alt.value(1), alt.value(0))
    )

    # Draw text labels near the points, and highlight based on selection
    text = line.mark_text(align="left", dx=5, dy=-5).encode(
        text=alt.condition(nearest, "nombre:Q", alt.value(" "))
    )

    # Draw a rule at the location of the selection
    rules = alt.Chart(names).mark_rule(color="gray").encode(
        x='annais:T'
    ).transform_filter(
        nearest
    )

    # Put the five layers into a chart and bind the data
    layers = alt.layer(
        line, selectors, points, rules, text
    ).properties(
        width=600, height=300
    )
    display(layers)
    
    


searchbar.on_submit(searchbar_callback)


```
